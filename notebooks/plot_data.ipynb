# Plotting synthetic data sets - one with anomalies and one without

We can generate new datasets with the code below if we want to.

```python
import sys, os
sys.path.append(os.path.abspath("..")) #need to do this to find src
from src.data.synthetic import generate_blob_dataset_with_anomalies, generate_blob_dataset

generate_blob_dataset(
        n_samples=500,
        n_features=2,
        n_clusters=2,
        cluster_std=0.8,
        to_csv="../data/synthetic_blobs.csv"
    )
print("Synthetic dataset saved to data/synthetic_blobs.csv")

generate_blob_dataset_with_anomalies(
        n_samples=1000,
        n_features=2,
        n_clusters=2,
        cluster_std=0.8,
        to_csv="../data/synthetic_blobs_anomalies.csv"
    )
print("Synthetic dataset with anomalies saved to data/synthetic_blobs_anomalies.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('../data/synthetic_blobs.csv')
plt.figure(figsize=(8, 6))
plt.scatter(df['feature_1'], df['feature_2'])
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot of X vs Y')
plt.show()
```
